```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def nmperh_ftmin(nmperh):
    return nmperh*6076/60

def nm_ft(nm):
    return nm*6076

def rad(deg):
    return deg*np.pi/180

def landing_d_rate(deg,vk):
    return nmperh_ftmin(vk*np.tan(rad(deg)))

def landing_ynm_deg(deg,ynm):
    return nm_ft(ynm*np.tan(rad(deg)))

def landing_xft_deg(deg,xft):
    ynm = (xft/np.tan(rad(deg)))/6076
    return ynm
```

```python
x = np.linspace(100,200)
y = landing_d_rate(3,x)
plt.xlabel("knots")
plt.ylabel("ft/min")
plt.grid()
plt.plot(x,y)

landing_d_rate(3,190)
```

```python
x=np.linspace(0,1000)
y = landing_xft_deg(3,x)
plt.grid()
plt.xlabel("nm")
plt.ylabel("ft")
plt.plot(y,x)
```
